# Check pre-experiment survey

This notebook explores participant demographics based on those who filled in the pre-experiment survey

```python
import pandas as pd
```

```python
raw = pd.read_csv("background_survey_labels.csv", header=None)
col_names = raw.iloc[2].apply(
    lambda x: x.split('"ImportId":"')[1].rstrip('"}') if '"ImportId":"' in str(x) else x
).tolist()
demo = raw.iloc[3:].copy()
demo.columns = col_names
demo = demo.reset_index(drop=True)

# participant IDs
demo["participant_id"] = demo["QID12_TEXT"].str.strip()
demo = demo.dropna(subset=["participant_id"]).reset_index(drop=True)

#remapping
pid_mapping = {
    "P007": "P008",
    "P038": "P040",
    "P056": "P057",
    "P070": "P072",
    "P074": "P075",
}
demo["participant_id"] = demo["participant_id"].replace(pid_mapping)

# filter to complete participants
complete_pids = ["P002", "P008", "P011", "P012", "P014", "P032", "P036", "P040",
                  "P045", "P047", "P053", "P057", "P061", "P066", "P067", "P068",
                  "P069", "P072", "P075", "P082"]
missing_presurvey = ["P011", "P047", "P068"]
valid_pids = [pid for pid in complete_pids if pid not in missing_presurvey]

demo = demo[demo["participant_id"].isin(valid_pids)].reset_index(drop=True)

print(f"N = {len(demo)} participants with demographic data")
print(demo["participant_id"].tolist())
```

## Check demographics

```python
print("Gender:")
print(demo["QID1"].value_counts())

print("\nAge Group:")
print(demo["QID4"].value_counts())

print("\nEnglish First Language:")
print(demo["QID2"].value_counts())

print("\nEducation Level:")
print(demo["QID7"].value_counts())

print("\nField of Work/Study:")
print(demo["QID6"].value_counts())

print("\nAI Chatbot Familiarity:")
print(demo["QID8"].value_counts())

print("\nAI Chatbot Usage Frequency:")
print(demo["QID9"].value_counts())

print("\nOverall Opinion of AI Chatbots:")
print(demo["QID10"].value_counts())
```
